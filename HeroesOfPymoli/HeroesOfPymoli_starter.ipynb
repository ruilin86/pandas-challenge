```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
len(purchase_data["SN"].unique())
total_players={len(purchase_data["SN"].unique())}
total_players=pd.DataFrame(total_players,columns=["Total"])
total_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
purchase_data.head()
```

```python
item_num=len(purchase_data["Item ID"].unique())
average_price=round(purchase_data["Price"].mean(),2)
purchase_num=purchase_data["Purchase ID"].count()
total_rev=purchase_data["Price"].sum()

summary=pd.DataFrame({"Number of Unique Items":[item_num],
                      "Average Price":[average_price],
                      "Number of Purchases":[purchase_num],
                      "Total Revenue": [total_rev]})

summary["Average Price"]=summary["Average Price"].map("${:,.2f}".format)
summary["Total Revenue"]=summary["Total Revenue"].map("${:,.2f}".format)
summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# select columns "SN" and "Gender"
player_gender=purchase_data[["SN","Gender"]]

# remove duplicates
player_gender=player_gender.drop_duplicates()

# calculate total # of players
total_player = player_gender["SN"].count()

# group bu gender
gender_count = player_gender.groupby(player_gender["Gender"])

# count # of players by gender and add the output into a dataframe
gender_count = gender_count.count()

# calculate the player percentage
gender_percent = gender_count["SN"]/total_player

# add a column to the dataframe
gender_count["Percentage of Player"] = gender_percent

#rename column "SN" to "Total Count"
gender_count = gender_count.rename(columns={"SN": "Total Count"})

# format number to percentage
gender_count["Percentage of Player"] = gender_count["Percentage of Player"].map("{:.2%}".format)

gender_count
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data.head()
```

```python
# group data by gender
df = purchase_data.groupby(purchase_data["Gender"])
purchase_count = df["Purchase ID"].count()
avg_purchase_price = round(df["Price"].mean(),2)
total_purchase_value = df["Price"].sum()
avg_purchase_person = round(total_purchase_value/gender_count["Total Count"],2)

purchase_gender_summary=pd.DataFrame({
    "Purchase Count":purchase_count,
    "Average Purchase Price": avg_purchase_price,
    "Total Purchase Value": total_purchase_value,
    "Average Purchase Per Person": avg_purchase_person
})

purchase_gender_summary["Average Purchase Price"]=purchase_gender_summary["Average Purchase Price"].map("${:,.2f}".format)
purchase_gender_summary["Total Purchase Value"]=purchase_gender_summary["Total Purchase Value"].map("${:,.2f}".format)
purchase_gender_summary["Average Purchase Per Person"]=purchase_gender_summary["Average Purchase Per Person"].map("${:,.2f}".format)
purchase_gender_summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins=[0,9.99,14.99,19.99,24.99,29.99,34.99,39.99,999]
group_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

player_age=purchase_data[["SN","Age"]].drop_duplicates()

player_age["Age Ranges"]=pd.cut(player_age["Age"], bins, labels=group_names)

player_age_group=player_age.groupby(player_age["Age Ranges"])

age_count=player_age_group["Age Ranges"].count()

age_percent=(player_age_group["Age Ranges"].count()/player_age["SN"].count()).map("{:.2%}".format)

age_count_summary=pd.DataFrame({"Total Count":age_count,
                        "Percentage": age_percent})                             
age_count_summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"],bins, labels=group_names)
age = purchase_data.groupby(purchase_data["Age Ranges"])
age
```

```python
count = age["Purchase ID"].count()
price = age["Price"].mean().map("${:.2f}".format)
value = age["Price"].sum()
avg_person = (value/age_count).map("${:,.2f}".format)

purchase_age = pd.DataFrame({
    "Purchase Count": count,
    "Average Purchase Price":price,
    "Total Purchase Value":value,
    "Avg Total Purchase per Person":avg_person
})

purchase_age["Total Purchase Value"] = purchase_age["Total Purchase Value"].map("${:,.2f}".format)
purchase_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_data.head()
```

```python
spender=purchase_data.groupby(purchase_data["SN"])
count=spender["Purchase ID"].count()
price=spender["Price"].mean().map("${:,.2f}".format)
value=spender["Price"].sum()
spender_summary = pd.DataFrame({
    "Purchase Count": count,
    "Average Purchase Price": price,
    "Total Purchase Value": value})
spender_summary = spender_summary.sort_values("Total Purchase Value", ascending = False)
spender_summary["Total Purchase Value"] = spender_summary["Total Purchase Value"].map("${:,.2f}".format)
spender_summary
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_data = purchase_data[["Item ID","Item Name","Price"]]
popular_data = popular_data.rename(columns={"Price":"Item Price"})
popular_group = popular_data.groupby(["Item ID","Item Name"])
count = popular_group["Item ID"].count()
price = popular_group["Item Price"].mean().map("${:,.2f}".format)
value = popular_group["Item Price"].sum()

pd = pd.DataFrame({
    "Purchase Count": count,
    "Item Price": price,
    "Total Purchase Value": value
})

popular_summary = pd.sort_values("Purchase Count", ascending = False)
popular_summary["Total Purchase Value"]=popular_summary["Total Purchase Value"].map("${:,.2f}".format)
popular_summary
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profit_summary=pd.sort_values("Total Purchase Value", ascending = False)
profit_summary["Total Purchase Value"] = profit_summary["Total Purchase Value"].map("${:,.2f}".format)
profit_summary
```
