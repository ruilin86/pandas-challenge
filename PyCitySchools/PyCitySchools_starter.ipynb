```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_school = len(school_data_complete["school_name"].unique())
total_student = len(school_data_complete["Student ID"].unique())
total_budget = school_data["budget"].sum()
average_math = school_data_complete["math_score"].mean()
average_reading = school_data_complete["reading_score"].mean()
math_passing_percent = school_data_complete.loc[school_data_complete["math_score"]>=70.0]["Student ID"].count()/total_student*100
reading_passing_percent = school_data_complete.loc[school_data_complete["reading_score"]>=70.0]["Student ID"].count()/total_student*100
overall_passing_percent = school_data_complete.loc[(school_data_complete["math_score"]>=70.0) & (school_data_complete["reading_score"]>=70.0)]["Student ID"].count()/total_student*100
```

```python
result_df=pd.DataFrame({
    "Total School": [total_school],
    "Total Student": [total_student],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math],
    "Average Reading Score": [average_reading],
    "% Passing Math":[math_passing_percent],
    "% Passing Reading":[reading_passing_percent],
    "% Overall Passing Rate": [overall_passing_percent]
})

result_df["Total Budget"]=result_df["Total Budget"].map("${:,.2f}".format)
result_df["Average Math Score"]=result_df["Average Math Score"].map("{:,.2f}".format)
result_df["Average Reading Score"]=result_df["Average Reading Score"].map("{:,.2f}".format)
result_df["% Passing Math"]=result_df["% Passing Math"].map("{:,.2f}".format)
result_df["% Passing Reading"]=result_df["% Passing Reading"].map("{:,.2f}".format)
result_df["% Overall Passing Rate"]=result_df["% Overall Passing Rate"].map("{:,.2f}".format)
result_df
```
